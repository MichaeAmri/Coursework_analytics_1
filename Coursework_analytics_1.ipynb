```python
# 1.0 Подготовка датасетов к работе

# объединяем progress_phases и progresses в main
main = pd.merge(
                pd.read_csv(
                            'Data/progress_phases.csv',
                            parse_dates=['start_date', 'finish_date'],
                            dayfirst=False, infer_datetime_format=True
                            ),
                pd.read_csv('Data/progresses.csv'),
                left_on='progress_id', right_on='id'
                ).drop('id', axis=1)

# добавляем courses
main = main.merge(
                    pd.read_csv('Data/courses.csv'),
                    left_on='course_id', right_on='id'
                    ).drop(['id', 'Unnamed: 0'], axis=1)

# добавляем course_contents
main = main.merge(
                    pd.read_csv('Data/course_contents.csv'),
                    left_on=['course_id', 'module_number', 'lesson_number'],
                    right_on=['course_id', 'module_number', 'lesson_number']
                    )

# меняем названия некоторых столбцов
main.rename(columns={'title':'course_title', 'field':'course_field'}, inplace=True) 

# проверяем основной датасет
main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf'].head(3)
```

```python
# 1.1.1 Описание данных: максимльный номер модуля, медианное количество уроков в модуле, количество учеников

# проходим по списку курсов, для каждого курса определяем
# максимльный номер модуля, медианное количество уроков в модуле, 
# количество учеников
# и записываем этоти числа в соответствующие столбцы

main['modules_number'] = 0
main['median_lessons_number'] = 0
main['students_count'] = 0

for course in main['course_title'].unique():

    # количество модулей в курсе
    max_value = main[main['course_title'] == course]['module_number'].max()
    main.loc[main['course_title'] == course, 'modules_number'] = max_value

    # количество учеников для каждого курса
    students_count = len(main[main['course_title'] == course]['student_id'].unique())
    main.loc[main['course_title'] == course, 'students_count'] = students_count

    # извлекаем количество уроков в каждом модуле в список
    # и считаем медиану по этому списку, записываем в соответствующий столбец
    lesson_numbers = []

    for module in main[main['course_title'] == course]['module_title'].unique():
        lesson_numbers.append(
        (main[(main['course_title'] == course) & (main['module_title'] == module)]
        .drop(['student_id', 'progress_id', 'start_date', 'finish_date', 'status'], axis=1)
        .reset_index(drop=True).drop_duplicates()
        .sort_values('module_title').shape[0]))

    main.loc[main['course_title'] == course, 'median_lessons_number'] = int(np.median(lesson_numbers))


info = (main.groupby(['course_title', 'modules_number', 'median_lessons_number', 'students_count', 'module_title'])
            .agg(max)[['module_number', 'lesson_number']]
            .sort_values(['course_title', 'module_number']))

print('Общее количество курсов в датасете: ', len(main['course_title'].unique()))
```

```python
# 1.1.2 Описание данных: возраст студентов
students = pd.read_csv('Data/students.csv', parse_dates=['birthday'], dayfirst=False)
students['birthday'] = pd.to_datetime(students['birthday'], yearfirst=True, dayfirst=False, errors='coerce')
students.dropna(inplace=True, subset=['birthday'])

students['Age'] = students.apply(lambda s: pd.to_datetime('today') - s['birthday'], axis=1)
students['Age'] = students['Age'].astype('<m8[Y]')
students.drop(students[students['Age'] < 18 ].index, inplace=True)

# датасет age_1 содержит обзую информацию о возрастах
age_1 = (students[['Age']].agg(['min', 'max', 'mean', 'median']))

# датасет age_2 содержит информацию о возрастах, сгруппированную по курсам
age_2 = (main[['course_title', 'student_id']]
.merge(students, left_on='student_id', right_on='id')
.groupby('course_title')
.agg(['min', 'max', 'mean', 'median'])[['Age']])
```

##### 1 **(p)** Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней работы.

```python
# формируем список дат, когда студенты начинали проходить курс.
# студент мог сделать первую домашнюю работу не из первого урока первого модуля,
# поэтому нужно найти минимальную дату начала занятий для каждого студента через .min()
start_dates = main.groupby(['course_title', 'student_id', 'progress_id'])[['start_date']].min()

# отбрасываем информацию о дне и времени начала, группируем по месяцам, затем считаем количество студентов
start_dates['start_date'] = start_dates.apply(lambda x: str(x.start_date.to_period('M')), axis=1)
start_dates.reset_index(inplace=True)
start_dates = start_dates.groupby(['start_date', 'course_title'])[['student_id']].count()
start_dates.reset_index(inplace=True)
start_dates.head()
```

```python
fig, axs = plt.subplots(figsize=(18, 72), nrows=len(main['course_title'].unique()), ncols=1)

# для каждого направления строим свой график
for i, val in enumerate(main['course_title'].unique()):
    sns.lineplot(data=start_dates[start_dates['course_title'] == val], x='start_date', y='student_id', ax=axs[i])
    axs[i].set_title(val, fontsize=16)
    axs[i].set_xlabel('')
    axs[i].set_ylabel('Прирост учеников')
    axs[i].set_xlabel('Месяц')

fig.suptitle("Прирост учеников по месяцам для каждого курса", y = 0.89, fontsize=18)
plt.subplots_adjust(hspace=0.4)
plt.box(on=None)
plt.show()
```

##### 4 **(p)** Рассчитать количество прогрессов по выполнению домашних работ в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно) для каждого курса. Учитывать, что выполнение домашнего задания может перетекать из одного месяца в другой (такие дз надо включать в общее число прогрессов для всех месяцев, которые покрывает срок выполнения этих дз)

```python
# интересны только столбцы course_title, progress_id, start_date, finish_date из датафрейма main
# незаконченные прогрессы со статусом start, wait и fail отбрасываем
data = main[['course_title', 'progress_id', 'start_date', 'finish_date']].dropna()

# отбрасываем значения времени и дня у столбцов start_date finish_date, чтобы можно было их группировать
data['start_date'] = data.apply(lambda x: str(x.start_date.to_period('M')), axis=1)
data['finish_date'] = data.apply(lambda x: str(x.finish_date.to_period('M')), axis=1)

# для каждого прогресса находим минимальное и максимальное значение даты, т. е. начало и конец
start_dates = data.groupby('progress_id').agg('min')[['start_date']]
finish_dates = data.groupby('progress_id').agg('max')[['finish_date']]

# вспомогательный датафрейм df хранит индекс прогресса, его начало и конец 
df = start_dates.merge(finish_dates, on='progress_id')

# объединяем в датафрейм название курса, id прогресса, дату начала и дату окончания из датафрейма df
data = data[['course_title', 'progress_id']].merge(df, left_on='progress_id', right_index=True).drop_duplicates()
data.head()
```

```python
# добавляем вспомогательный столбец-ключ
data['number'] = 1

# формируем датасет со всеми месяцами, когда начинались или заканчивались прогрессы,
# в него также добавляем столбец-ключ
tmp = pd.DataFrame({'month':list(set(data['start_date']).union(set(data['finish_date'])))})
tmp.sort_values('month', inplace=True)
tmp['number'] = 1

# объединяем датасеты data и tmp по ключу number.
# теперь для каждого прогресса можно найти строку с любым из возможных месяцев
data = data.merge(tmp, on='number', how='right').drop(['number'], axis=1)
data.head()
```

```python
# в столбец is_in_interval записываем True, если временной диапазон прогресса содержит месяц из строки
data['is_in_interval'] = data.apply(lambda x: x.start_date <= x.month <= x.finish_date, axis=1)
```

```python
# Для каждого курса в каждом месяце считаем количество прогрессов
# месяцы без прогрессов удаляем
data = data.groupby(['month', 'course_title'])[['is_in_interval']].sum()
data.drop(data[data['is_in_interval'] == False].index, inplace=True)
data.reset_index(inplace=True)
data.head(50)
```

##### 1 **(p)** Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля (разность между временем начала и окончания выполнения домашней работы) для каждого курса. Если домашних заданий в модуле несколько, то считать разность между временем начала выполнения первой домашней работы и временем окончания выполнения последней домашней работы в модуле

```python
# берем только необходимые столбцы из датафрейма main
data = main[['course_title', 'module_number', 'progress_id', 'status', 'start_date', 'finish_date']]
   
# группируем данные по курсу, номеру модуля и id прогресса
# для каждой группы считаем количество уникальных статусов (done, start, fail или wait)
df = data.groupby(['course_title', 'module_number', 'progress_id'])[['status']].nunique()

# отбрасываем все записи, где статусов больше двух - интересуют только законченные модули
# в датафрейм могут попасть записи, где только 1 статус, но не done. это поправим дальше
df.drop(df[df['status'] > 1].index, inplace=True)

# путем объединения с датафреймом data восстанавливаем значения status, start date, finish date
# лишние столбцы удаляем
df.reset_index(inplace=True)
df = df.merge(data, on=['course_title', 'module_number', 'progress_id'], suffixes=('_num', '_val')).drop(columns='status_num')
df.drop(df[df['status_val'] != 'done'].index, inplace=True)
df.drop(columns='status_val', inplace=True)

# получаем датафрейм, в котором содержатся даты начала и завершения каждого прогресса с названием курса и номером модуля
df.head()
```

```python
# для каждого прогресса находим минимальную дату - дату начала ДЗ и максимальную дату - дату завершения ДЗ
df_grouped = df.groupby(['course_title', 'module_number', 'progress_id']).agg(['min', 'max'])
df_grouped.drop(columns=[('start_date', 'max'), ('finish_date', 'min')], inplace=True)
df_grouped = df_grouped.droplevel(0, axis=1).reset_index()

# в столбец duration записываем продолжительность выполнения ДЗ в часах
df_grouped['duration'] = (df_grouped['max'] - df_grouped['min']) / np.timedelta64(1, 'h')

# группируем и агрегируем датафрейм для получения требуемых данных для каждого курса каждого модуля
df_grouped = df_grouped.groupby(['course_title', 'module_number'])[['duration']].agg(['min', 'max', 'mean', 'median'])
df_grouped = df_grouped.droplevel(0, axis=1).reset_index()
df_grouped.head()
```

##### 3 **(p)**  Чтобы выявить сезонность, посчитать медианное время выполнения домашней работы по месяцам (12 месяцев, январь-декабрь) для каждого курса.

```python
# формируем датафрейм, в котором содержатся прогрессы с законченным домашним заданием и временем его выполнения
data = main[(main['is_homework'] == True) & (main['status'] == 'done')][[
    'course_title', 'progress_id', 'status', 'start_date', 'finish_date', 'is_homework']]
data['duration'] = data.apply(lambda x: (x.finish_date - x.start_date) / np.timedelta64(1, 'h'), axis=1)
data.drop(columns=['is_homework', 'status'], inplace=True)

# На каждый прогресс добавляем по 12 строк с месяцами от 1 до 12, результат сохраняем в df_month
months = pd.DataFrame({'month': range(1, 13)})
months['key'] = 1
data['key'] = 1
df_months = data.merge(months, on='key', how='right')

# если месяц не входит во временной диапазон выполнения дз, то такую строку удаляем
df_months['is_in'] = df_months.apply(lambda x: x.start_date.month <= x.month <= x.finish_date.month, axis=1)
df_months.drop(df_months[~df_months['is_in']].index, inplace=True)
df_months.drop(columns=['key', 'is_in'], inplace=True)

# группируем df_months по курсу и месяцу, считаем медианную продолжительность выполнения дз
df_month = df_months.groupby(['course_title', 'month'])[['duration']].mean()
df_month.head()
```

## 4 Расчет конверсии

#### 1 **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
# формируем датафреймы df_prev и df_current с номерами уроков первой и последней домашней работы соответственно
df = main[main['is_homework'] == True][['course_title', 'module_number', 'lesson_number', 'student_id']]
df_prev = df.groupby(['course_title', 'module_number'], as_index=False)[['lesson_number']].max()
df_current = df.groupby(['course_title', 'module_number'])[['lesson_number']].min()

# К каждому уроку присоединяем список студентов, которые этот урок прошли
# датафрейм df_prev должен содержать только законченные работы, т. е. работы со статусом "done"
df_prev = df_prev.merge(
    main[main['status'] == 'done'][['course_title', 'module_number', 'lesson_number', 'student_id']],
    on=['course_title', 'module_number', 'lesson_number'])
df_current = df_current.merge(
    main[main['status'] == 'start'][['course_title', 'module_number', 'lesson_number', 'student_id']],
    on=['course_title', 'module_number', 'lesson_number'])

# для каждого датафрейма считаем количество студентов
df_prev = df_prev.groupby(['course_title', 'module_number'])[['student_id']].count()
df_current = df_current.groupby(['course_title', 'module_number'])[['student_id']].count()

# соединяем датафреймы со смещением
df_funnel = df_current.merge(df_prev.shift(periods=1, fill_value=0), on=['course_title', 'module_number'], suffixes=['_start_cur', '_finish_prev']).reset_index()

# находим первый модуль с ДЗ и добавляем его номер к основному датафрейму
start_module_number = df_funnel.groupby('course_title')[['module_number']].min()
df_funnel = df_funnel.merge(start_module_number, on='course_title', suffixes=['', '_start'])

# функция для расчета конверсии
def conversion(x):
    if x.module_number != x.module_number_start:
        return x.student_id_start_cur / x.student_id_finish_prev
    else:
        return 1

# расчет конверсии
df_funnel['conversion'] = df_funnel.apply(lambda x: conversion(x), axis=1)
df_funnel.drop(columns=['student_id_start_cur', 'student_id_finish_prev', 'module_number_start'], inplace=True)
df_funnel
```

```python
rows_num = main['course_title'].nunique()
fig, axs = plt.subplots(figsize = (18, 6 * rows_num), nrows=rows_num, ncols=1)

for index, name in enumerate(main['course_title'].unique()):
    sns.barplot(data=df_funnel[df_funnel['course_title'] == name], x='module_number', y='conversion', ax=axs[index])
    axs[index].set_title(name)
    axs[index].set_xlabel('Номер модуля')
    axs[index].set_ylabel('Конверсия')
    axs[index].grid(axis='y')
```
